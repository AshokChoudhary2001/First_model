```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('dataset/train.csv')
```

```python
df.head(15)
```

```python
df.isnull().sum()
```

```python
 df.drop(columns = ['body' , 'boat' , 'cabin','name','ticket'],axis = 1 ,inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(['survived'],axis = 1) , df['survived'] , test_size = 0.2 , random_state= 42)

x_train.shape , x_test.shape ,y_train.shape
```

```python
x_train.isnull().sum()
```

## Simple imputer for filling all the missing values 

```python

si_age = SimpleImputer()
si_home = SimpleImputer(strategy= 'most_frequent')

x_train_age_fare = si_age.fit_transform(x_train[['age','fare']]) 
x_train_emb_home = si_home.fit_transform(x_train[['embarked' , 'home.dest']])

x_test_age_fare = si_age.transform(x_test[['age','fare']])
x_test_emb_home = si_home.transform(x_test[['embarked' , 'home.dest']])

x_train_age_fare , x_train_emb_home
```

## One hot encoding for the conversion of nominal data into numerical 


```python
df
```

```python
df.isnull().sum()
```

```python
df['home.dest'].value_counts()
```

**so we will not use x_Train dataframe for 'home.dest' and 'embarked'  as it contain missing values we will use the imputed ones**

```python
ohe_home_emb = OneHotEncoder(sparse_output = False  , handle_unknown = 'ignore' )
```

```python
#these is just for the  home.dest and embarked we have to create another object for sex column 
train_simple = ohe_home_emb.fit_transform( x_train_emb_home)
test_simple = ohe_home_emb.transform(x_test_emb_home)
test_simple.shape , train_simple.shape 
```

```python
ohe_sex = OneHotEncoder(sparse_output= False , handle_unknown = 'ignore')
```

```python
x_train.head()
```

```python
train_sex = ohe_sex.fit_transform(x_train[['sex']])
test_sex = ohe_sex.transform(x_test[['sex']])

train_sex.shape
```

```python
x_semi_train = x_train.drop(['sex' , 'embarked' ,'home.dest'] ,axis = 1) 
x_semi_test = x_test.drop(['sex' , 'embarked' , 'home.dest'] , axis = 1)

x_semi_train.shape 
```

```python
x_final_train_transformed = np.concatenate((x_semi_train,train_sex,train_simple),axis = 1) 
x_final_test_transformed = np.concatenate((x_semi_test,test_sex,test_simple) ,axis = 1 )
```

```python
x_final_train_transformed.shape , x_final_test_transformed.shape
```

##  nOW we will use our classifier 

```python
clf = DecisionTreeClassifier()
clf.fit(x_final_train_transformed,y_train)
```

```python
y_pred = clf.predict(x_final_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score 
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc)
```
